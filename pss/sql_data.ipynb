```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# x축에 표시할 연도와 시각화할 4개 항목 정의
years = df['년월'].tolist()
categories = ['승용합계', '승합합계', '화물합계', '특수합계']

x = np.arange(len(years))  # 연도별 위치 (0, 1, 2)
width = 0.1 # 개별 막대의 너비

fig, ax = plt.subplots(figsize=(24, 7))

# 4개의 데이터를 x축 중심으로부터 오프셋을 주어 배치
rects1 = ax.bar(x - 1.5*width, df['승용합계'], width, label='승용', color='royalblue')
rects2 = ax.bar(x - 0.5*width, df['승합합계'], width, label='승합', color='orange')
rects3 = ax.bar(x + 0.5*width, df['화물합계'], width, label='화물', color='forestgreen')
rects4 = ax.bar(x + 1.5*width, df['특수합계'], width, label='특수', color='crimson')

# 그래프 라벨 및 제목 설정
ax.set_ylabel('등록 대수')
ax.set_title('연도별 차종 그룹별 등록 현황 (4개 항목 비교)')
ax.set_xticks(x)
ax.set_xticklabels(years, rotation=45, ha='right')
ax.legend()

# 막대 위에 수치 표시 함수
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{int(height):,}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3), textcoords="offset points",
                    ha='center', va='bottom', fontsize=9)

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
autolabel(rects4)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 1. 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 2. 데이터 불러오기 (전체 데이터 사용)
# df = pd.read_csv('pj2601_연도별등록합계.csv')

# # 3. 합계 계산 (모든 행 대상)
# df['승용합계'] = df[['승용관용', '승용자가용', '승용영업용']].sum(axis=1)
# df['승합합계'] = df[['승합관용', '승합자가용', '승합영업용']].sum(axis=1)
# df['화물합계'] = df[['화물관용', '화물자가용', '화물영업용']].sum(axis=1)
# df['특수합계'] = df[['특수관용', '특수자가용', '특수영업용']].sum(axis=1)

categories = ['승용합계', '승합합계', '화물합계', '특수합계']
years = df['년월'].tolist() # 모든 행의 라벨을 리스트로 생성

# 4. 증감률 계산 (%)
df_pct = df[categories].pct_change() * 100

# 5. 시각화 설정
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 12))

# --- 상단: 전체 수치 추이 (Line Chart) ---
for cat in categories:
    ax1.plot(years, df[cat], marker='o', label=cat.replace('년월', ''), linewidth=2)
    for i, val in enumerate(df[cat]):
        ax1.text(i, val, f'{int(val):,}', ha='center', va='bottom', fontsize=9)

ax1.set_title('차종별 등록 대수 전체 추이', fontsize=15)
ax1.set_xticks(range(len(years))) # 모든 행에 틱 설정
ax1.set_xticklabels(years, rotation=45, ha='right') # 모든 행 라벨 표시
ax1.legend()
ax1.grid(True, linestyle='--', alpha=0.5)

# --- 하단: 전체 증감률 변화 (Bar Chart) ---
x_pct_labels = years[1:] # 첫 행 제외한 모든 라벨
x_indexes = np.arange(len(x_pct_labels))
width = 0.2
offsets = np.linspace(-width*1.5, width*1.5, 4)

for i, cat in enumerate(categories):
    vals = df_pct[cat].dropna()
    ax2.bar(x_indexes + offsets[i], vals, width, label=cat.replace('년월', ''))
    for j, val in enumerate(vals):
        color = 'red' if val > 0 else 'blue'
        ax2.text(j + offsets[i], val, f'{val:+.1f}%', ha='center', 
                va='bottom' if val > 0 else 'top', fontsize=8, color=color)

ax2.axhline(0, color='black', linewidth=1)
ax2.set_title('전기 대비 증감률 변화 (%)', fontsize=15)
ax2.set_xticks(x_indexes)
ax2.set_xticklabels(x_pct_labels, rotation=45, ha='right')
ax2.legend()

plt.tight_layout()
plt.show()
```
