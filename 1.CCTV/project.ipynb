```python
corr_df = df1[[
    '발생건수 (건)',
    '부상자수 (명)',
    '사망자수 (명)',
    '자동차 1만대당 발생건수 (건)',
    '인구 10만명당 사망자수 (명)',
    '인구 10만명당 부상자수 (명)'
]]
```

```python
corr_df.corr()
```

```python
cols = ['발생건수 (건)', '부상자수 (명)']

for c in cols:
    df1[c] = (
        df1[c]
        .str.replace(',', '', regex=False)
        .astype(int)
    )
```

```python
df1['사고당 사망률'] = df1['사망자수 (명)'] / df1['발생건수 (건)']
```

```python
df1['사고당 부상률'] = df1['부상자수 (명)'] / df1['발생건수 (건)']
```
