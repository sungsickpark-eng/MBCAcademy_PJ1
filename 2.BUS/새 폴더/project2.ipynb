```python
import pandas as pd
from pathlib import Path

DATA_DIR = Path(r"C:\minipj\참고자료")
files = sorted(DATA_DIR.glob("*.csv"))

dfs = []
for f in files:
    df = pd.read_csv(
        f,
        encoding="cp949",
        sep=",",
        engine="python",
        quotechar='"'
    )
    df["source_file"] = f.name
    dfs.append(df)

df_all = pd.concat(dfs, ignore_index=True)

print(df_all.shape)
print(df_all.head())
```

```python
df_all.to_csv(
    r"C:\minipj\참고자료\bus_all2.csv",
    index=False,
    encoding="cp949"
)
```
